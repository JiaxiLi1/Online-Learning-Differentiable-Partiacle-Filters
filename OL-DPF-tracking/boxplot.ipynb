```python
def set_colours(bp, box_color, linewidth, flier_color='purple', markersize=2, linestyle=None):
    for box in bp['boxes']:
        # change outline color
        box.set(color=box_color, linewidth=linewidth)
        if linestyle:
            box.set(linestyle=linestyle)
        # change fill color
        box.set(facecolor='white')

    # change color and linewidth of the whiskers
    for whisker in bp['whiskers']:
        whisker.set(color=box_color, linewidth=linewidth)

    # change color and linewidth of the caps
    for cap in bp['caps']:
        cap.set(color=box_color, linewidth=linewidth)
        if linestyle:
            cap.set(linestyle=linestyle)

    # change color and linewidth of the medians
    for median in bp['medians']:
        median.set(color=box_color, linewidth=linewidth)
        if linestyle:
            median.set(linestyle=linestyle)

    # change the style of fliers and their fill
    for flier in bp['fliers']:
        flier.set(marker='o', alpha=1.0)
        flier.set_markerfacecolor('None')
        flier.set_markeredgecolor(flier_color)
        flier.set_markersize(markersize)
# Updated plot_boxplot function with new figure size (10, 6)
# Updated plot_boxplot function to center the boxplots around the specified time steps
def plot_boxplot(dataset1, dataset2, dataset3, label1, label2, label3):
    linewidth = 2
    markersize = 3
    box_width = 0.1
    offset = 0.125
    labels = [label1, label2, label3]
    colors = [ 'red','blue','green']
    tick_params_size = 22
    axis_label_size = 22
    evaluated_steps = [1, 1000, 2000, 3000, 4000, 5000]
    data_list = [dataset1[:, np.array(evaluated_steps) - 1], dataset2[:, np.array(evaluated_steps) - 1], dataset3[:, np.array(evaluated_steps) - 1]]
    bp_list = []

    fig = plt.figure(1, figsize=(14, 6))
    ax = fig.add_subplot(111)

    for j, time_step in enumerate(evaluated_steps):
        for i, data in enumerate(data_list):
            central_position = j + offset * (i - 1)
            bp = plt.boxplot(data[:, j], positions=[central_position], patch_artist=True, widths=box_width)
            set_colours(bp, colors[i], linewidth, markersize=markersize)
            if j == 0:
                bp_list.append(bp["boxes"][0])

    plt.xticks(np.array(range(len(evaluated_steps))), evaluated_steps)
    plt.xlabel('Time step', fontsize=axis_label_size)
    plt.tick_params(direction='in', labelsize=tick_params_size)
    plt.ylabel('RMSE', fontsize=axis_label_size)
    plt.legend(bp_list, labels, loc='upper right', fontsize=22, ncol=1)
    plt.grid(alpha=0.2)
    # plt.title('Validation RMSE at evaluated time steps', fontsize=axis_label_size)
    plt.savefig(savepath, bbox_inches='tight', pad_inches=0.05)
    plt.show()


# Plot the boxplots with the updated figure size
print(all_loss_recorder1.shape, all_loss_recorder2.shape)
plot_boxplot(all_loss_recorder1, all_loss_recorder2, all_loss_recorder3, "Online learning","Pre-training","Supervised learning")
```
