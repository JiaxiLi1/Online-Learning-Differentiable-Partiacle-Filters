```python
def normalize_log_probs(probs, dim=2):
    probs_max=probs.max(axis=dim, keepdims=True)[0]
    probs_minus_max = probs-probs_max
    probs_normalized = np.exp(probs_minus_max)
    probs_normalized = probs_normalized / np.sum(probs_normalized, axis=dim, keepdims=True)
    return probs_normalized
def fun(x, mean=0, cov=1):
    return scipy.stats.multivariate_normal.pdf(x, mean=mean, cov=cov)

def mark_inset(parent_axes, inset_axes, loc1a=1, loc1b=1, loc2a=2, loc2b=2, **kwargs):
    rect = TransformedBbox(inset_axes.viewLim, parent_axes.transData)

    pp = BboxPatch(rect, fill=False, **kwargs)
    parent_axes.add_patch(pp)

#     p1 = BboxConnector(inset_axes.bbox, rect, loc1=loc1a, loc2=loc1b, **kwargs)
#     inset_axes.add_patch(p1)
#     p1.set_clip_on(False)
#     p2 = BboxConnector(inset_axes.bbox, rect, loc1=loc2a, loc2=loc2b, **kwargs)
#     inset_axes.add_patch(p2)
#     p2.set_clip_on(False)

    return pp,# p1, p2

def lki_data(measurement_model_, 
             path='./logs/first_collected_result/vanilla_cnf/1_NF^False_DPF_20.0_20.0_2.5_0.0001_resample^ot_CNF',
             lki_type='cos'):
    hidden_size = 32
    state_dim = 2
    n_sequence = 2
    
    ckpt_e2e = torch.load(os.path.join(path,'models/e2e_model_bestval_e2e.pth'))
    
    cnf_measurement = build_conditional_nf(n_sequence, hidden_size, hidden_size, init_var=0.1, prior_std=10.0).to(device)
    likelihood_est = build_likelihood(hidden_size, state_dim).to(device)
    particle_encoder = build_particle_encoder(hidden_size , state_dim).to(device)
    encoder=build_encoder(hidden_size).to(device)
    
    cnf_measurement.load_state_dict(ckpt_e2e['cnf_measurement'])
    likelihood_est.load_state_dict(ckpt_e2e['likelihood_est'])
    particle_encoder.load_state_dict(ckpt_e2e['particle_encoder'])
    encoder.load_state_dict(ckpt_e2e['encoder'])

    result = np.load(os.path.join('./logs/first_collected_result/vanilla_NN/3_NF^False_DPF_20.0_20.0_2.5_0.0001_resample^ot_NN','data/test_result.npz'))
    normalized_likelihood = normalize_log_probs(result['likelihood_list'])
    obs = result['images']
    states=result['state'][..., :2]
    particles=result['particle_list']
    
    if lki_type=='cos':
        measurement_model = measurement_model_(particle_encoder)
    elif lki_type=='CNF':
        measurement_model = measurement_model_(particle_encoder, cnf_measurement)
    elif lki_type=='NN':
        measurement_model = measurement_model_(particle_encoder, likelihood_est)

    lki_recorder_log=torch.zeros(obs.shape[:2])
    with torch.no_grad():
        for i in range(obs.shape[0]):
            encodings = encoder(torch.tensor(obs[i]).to(device))
            lki_log = measurement_model(encodings, torch.zeros([obs.shape[1],1,2]).to(device))
            #print(lki_log)
            lki_recorder_log[i]=lki_log[:,0]
    lki_recorder = normalize_log_probs(lki_recorder_log.cpu().numpy(), dim=1)
    lki_recorder /= lki_recorder.max()
    
    return states, lki_recorder

def plot_likelihood_surface(states, lki_recorder, cov=50.0, plot_gaussian=False):
    fig = plt.figure(figsize=(24, 12))
    s=15
    n_batch=1
    xy_plane_width = 80
    z_height = 1.0
    alpha=0.75
    gaussian_range=85

    ax = fig.add_subplot(121,projection='3d')
    gaussian_range=80
    split=1/10
    space=5

    x_left = np.arange(-gaussian_range, -gaussian_range*split, space)
    y_left = np.arange(-gaussian_range, gaussian_range, space)
    X_left, Y_left = np.meshgrid(x_left, y_left)

    x_upper = np.arange(-gaussian_range*split, gaussian_range,space)
    y_upper = np.arange(gaussian_range*split, gaussian_range, space)
    X_upper, Y_upper = np.meshgrid(x_upper, y_upper)

    x_right = np.arange(gaussian_range*split, gaussian_range, space)
    y_right = np.arange(-gaussian_range*split, gaussian_range*split, space)
    X_right, Y_right = np.meshgrid(x_right, y_right)

    x_bottom = np.arange(-gaussian_range*split, gaussian_range, space)
    y_bottom= np.arange(-gaussian_range, -gaussian_range*split, space)
    X_bottom, Y_bottom = np.meshgrid(x_bottom, y_bottom)

    x_middle = np.arange(-gaussian_range*split, gaussian_range*split, 1)
    y_middle = np.arange(-gaussian_range*split, gaussian_range*split, 1)
    X_middle, Y_middle = np.meshgrid(x_middle, y_middle)

    data=np.concatenate([np.concatenate([X_upper.ravel()[:,None], Y_upper.ravel()[:,None]],axis=-1),
                         np.concatenate([X_left.ravel()[:,None], Y_left.ravel()[:,None]],axis=-1),
                         np.concatenate([X_right.ravel()[:,None], Y_right.ravel()[:,None]],axis=-1),
                         np.concatenate([X_bottom.ravel()[:,None], Y_bottom.ravel()[:,None]],axis=-1),
                         np.concatenate([X_middle.ravel()[:,None], Y_middle.ravel()[:,None]],axis=-1)], axis=0)
    data=np.concatenate([np.concatenate([X_middle.ravel()[:,None], Y_middle.ravel()[:,None]],axis=-1)], axis=0)
    zs = fun(data, mean=[0,0], cov=[[cov,0.], [0., cov]])
    Z=(1/zs.max())*zs.reshape(data.shape[0])
    if plot_gaussian:
        ax.scatter(data[:,0], data[:,1], Z, c=zs.reshape(data.shape[0]), s=s, cmap='jet',alpha=0.5)

    ax.scatter(states.reshape([-1,2])[:,0], states.reshape([-1,2])[:,1], lki_recorder.reshape([-1]),
              c=lki_recorder.reshape([-1]), cmap='jet', alpha=alpha, s=s)

    # ax.scatter(diff_cnf[:n_batch,:].reshape([-1,2])[..., 0], diff_cnf[:n_batch,:].reshape([-1,2])[..., 1],
    #            normalized_likelihood_cnf[:n_batch,:].reshape([-1]),
    #            c=normalized_likelihood_cnf[:n_batch,:].reshape([-1]), cmap='jet', alpha=alpha, s=s)
    #ax.set_zlim(0.0, z_height)
    ax.set_xlim(-xy_plane_width, xy_plane_width)
    ax.set_ylim(-xy_plane_width, xy_plane_width)
```

```python
import numpy as np
import os
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import MaxNLocator
from scipy.stats import sem, t
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
def load_data(folder):
    eval_loss_epoch=np.load(os.path.join(folder,'data','eval_loss_epoch.npy'))
    eval_prediction_best=np.load(os.path.join(folder,'data','eval_result_best.npz'))
    test_loss=np.load(os.path.join(folder,'data','test_loss_epoch.npy'))
    test_prediction=np.load(os.path.join(folder,'data','test_result.npz'))
    return eval_loss_epoch, eval_prediction_best, test_loss, test_prediction

def load_multiple_runs(folder):
    save_folders=os.listdir(folder)
    list_data=[]
    for data_folder in save_folders:
        list_data.append(load_data(os.path.join(folder,data_folder)))
    return list_data

def concatenate_eval_loss(list_data):
    loss_all=np.expand_dims(list_data[0][0],axis=0)

    for data in list_data:
        loss_all=np.concatenate([loss_all,np.expand_dims(data[0],axis=0)])
    return loss_all

def concatenate_test_loss(list_data):
    loss_all=np.expand_dims(list_data[0][2],axis=0)
    for data in list_data:
        loss_all=np.concatenate([loss_all,np.expand_dims(data[2],axis=0)])
    return loss_all

def concatenate_rmse_best_eval(list_data,batch_size=50,length=50):
    rmse_all=np.zeros([len(list_data),batch_size, length])
    for i,data in enumerate(list_data):
        rmse_all[i]=((data[1]['pred']-data[1]['state'][:,:,:2])**2).sum(axis=-1)**0.5
    return rmse_all

def concatenate_rmse_test(list_data,batch_size=50,length=50):
    rmse_all=np.zeros([len(list_data),batch_size, length])
    for i,data in enumerate(list_data):
        rmse_all[i]=((data[-1]['pred']-data[-1]['state'][:,:,:2])**2).sum(axis=-1)**0.5
    return rmse_all

def concatenate_ESS_best_eval(list_data, batch_size=50, length=50):
    ESS_all=np.zeros([len(list_data),batch_size, length])
    for i,data in enumerate(list_data):
        ESS_all[i]=1/np.sum(data[1]['particle_weight_list']**2, axis=-1)
    return ESS_all

def concatenate_ESS_test(list_data, batch_size=50, length=50):
    ESS_all=np.zeros([len(list_data),batch_size, length])
    for i,data in enumerate(list_data):
        ESS_all[i]=1/np.sum(data[-1]['particle_weight_list']**2, axis=-1)
    return ESS_all

def concatenate_likelihood_test(list_data, batch_size=50, length=50):
    ESS_all=np.zeros([len(list_data),batch_size, length])
    for i,data in enumerate(list_data):
        ESS_all[i]=1/np.sum(data[-1]['particle_weight_list']**2, axis=-1)
    return ESS_all

def set_colours(bp, box_color,flier_color='purple',markersize=2,linestyle=None):
    for box in bp['boxes']:
        # change outline color
        box.set( color=box_color, linewidth=linewidth, linestyle=linestyle)
        # change fill color
        box.set( facecolor = 'white' )

    ## change color and linewidth of the whiskers
    for whisker in bp['whiskers']:
        whisker.set(color=box_color, linewidth=linewidth)

    ## change color and linewidth of the caps
    for cap in bp['caps']:
        cap.set(color=box_color, linewidth=linewidth, linestyle=linestyle)

    ## change color and linewidth of the medians
    for median in bp['medians']:
        median.set(color=box_color, linewidth=linewidth, linestyle=linestyle)

    ## change the style of fliers and their fill
    for flier in bp['fliers']:
        flier.set(marker='o',alpha=1.0)
        flier.set_markerfacecolor('None')
        flier.set_markeredgecolor(flier_color)
        flier.set_markersize(markersize)
        
def all_data(setup_folder):
    list_data=load_multiple_runs(setup_folder)
    baseline_eval_loss=concatenate_eval_loss(list_data)
    baseline_rmse_best_eval=concatenate_rmse_best_eval(list_data)
    baseline_test_loss=concatenate_test_loss(list_data)
    baseline_rmse_test=concatenate_rmse_test(list_data)
    baseline_ESS_best_eval=concatenate_ESS_best_eval(list_data)
    baseline_ESS_test=concatenate_ESS_test(list_data)
    return baseline_eval_loss,baseline_rmse_best_eval,baseline_test_loss,baseline_rmse_test,baseline_ESS_best_eval,baseline_ESS_test

def rmse_evaluated_steps(setup_rmse,evaluated_step=list(np.arange(4,50,5))):
    evaluated_step.insert(0,0)
    evaluated_step=np.array(evaluated_step)
    setup_rmse_evaluated_steps=[]
    for i,step in enumerate(evaluated_steps):
        setup_rmse_evaluated_steps.append(setup_rmse.reshape([-1,50])[:,step])
        #setup_rmse_evaluated_steps.append(setup_rmse.mean(0)[:,step])
#         if i!=0:
#             setup_rmse_evaluated_steps[i]*=1.2
    return setup_rmse_evaluated_steps
```
