```python
# mean with confidence interval
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem, t

# Load data
# mean1 = np.load('mean1.npy')
# mean2 = np.load('mean2.npy')
# mean3 = np.load('mean3.npy')
# std1 = np.load('std1.npy')
# std2 = np.load('std2.npy')
# std3 = np.load('std3.npy')

# Confidence level
confidence = 0.95

# Lists to store means and standard deviations
means = [mean1, mean2]
stds = [std1, std2]

labels = ["mul", "sys"]
colors = ['blue', 'red']

# Plotting
plt.figure(figsize=(10, 8))
for i, (mean, std) in enumerate(zip(means, stds)):
    n = mean.shape[0]
    m = np.mean(mean)
    std_err = sem(mean)
    conf_interval = std_err * t.ppf((1 + confidence) / 2, n - 1)
    plt.plot(np.arange(1, 101), mean, label=labels[i], color=colors[i])
    plt.fill_between(np.arange(1, 101), mean - conf_interval, mean + conf_interval, alpha=0.15)

plt.legend(loc='upper right', fontsize=28, ncol=1)
plt.tick_params(direction='in', labelsize=28)
plt.grid(alpha=0.2)
plt.ylabel('RMSE', fontsize=28)
plt.xlabel('Epoch', fontsize=28)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
# plt.xticks(list(range(0, 51, 10)))
# plt.yticks(range(0, 30, 5))
# plt.title('RMSE error on validation set', fontsize=20)
plt.savefig(savepath, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
